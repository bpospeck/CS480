# Overview

This assignment will take a set of data about an airfoil's self-noise and partition the data into 3 disjoint sets: Train, validate, and test. These will then be used to perform linear ridge regression. The objective of ridge regression is to minimize the sum of squared errors as well as reduce the sum of the weights' magnitudes:
$$ \sum_{i=1}^N (\tv_i - \xv_i^T \wv)^2 + \lambda \sum_{i=2}^N w_i^2$$

Where $\tv$ is the target, $\xv$ is the input, $\wv$ is the weight, and $\lambda$ is a chosen constant. When $\lambda = 0$, we have a standard linear regression. $\lambda$ is used in order to essentially add in a cost for the magnitudes of the weights. 

In order to find $\wv$, the following equation will be used:

$$ \wv = (X^T X + \lambda I)^{-1} X^T T $$

Normally, $\lambda I$ will be a perfectly diagonal matrix like this:

$$
    \lambda I =
     \begin{bmatrix}
     \lambda & 0 & \dotsc & 0\\
       0 & \lambda & \dotsc & 0\\
       \vdots \\
       0 & 0 & \dotsc & \lambda
    \end{bmatrix}
$$

But since the first entry in the matrix will be based on a bias weight, we do not want to include a penalty for that weight. Thus, it will actually look like this:

$$
    \lambda I =
     \begin{bmatrix}
     0 & 0 & \dotsc & 0\\
       0 & \lambda & \dotsc & 0\\
       \vdots \\
       0 & 0 & \dotsc & \lambda
    \end{bmatrix}
$$

In order to find the best value of $\lambda$, we will be finding validation set errors in a number of folds with different values of $\lambda$. The lowest error produced in each fold will be returned and that will tell us which $\lambda$ is the best to use for the data.

# Method

The `train` function takes all attribute inputs, the targets, and a single value for $\lambdav$. It standardizes the inputs, `X`, before performing ridge regression. `train` returns a `model` in the form of a dictionary with 3 entries:
    1. 'means': The means of the inputs
    2. 'stds' : The standard deviations of the inputs
    3. 'w'    : The associated weights of the input attributes

```python
def train(X,T,lamb):
    """Standardizes the inputs, X, before calculating the means, standard deviations, 
and weights which are all returned in a dictionary. Performs ridge regression."""
    mean = np.mean(X,axis=0)
    std = np.std(X,axis=0)
    # below for loop standardizes the inputs before calculating w
    cols = X.shape[1]
    Z = np.empty(X.shape)
    for c in range(cols):
        Z[:,c] = (X[:,c]-mean[c]) / std[c]
    # Now insert column of 1's in front to account for w0
    Z1 = np.insert(Z, 0, 1, axis=1)
    w = np.linalg.lstsq(np.dot(Z1.T,Z1 + lamb),np.dot(Z1.T,T))[0]
    return {'means': mean, 'stds': std, 'w': w}
```

The `use` function takes the output `model` from `train` along with the inputs, `X`. It once again standardizes the inputs and returns a prediction of the target values in matrix form.

```python
def use(model,X):
    """Uses the returned dictionary from the train function in order to calculate 
and return the predicted matrix"""
    Z = np.empty(X.shape)
    # Standardize X again
    cols = X.shape[1]
    for c in range(cols):
        Z[:,c] = (X[:,c] - model['means'][c]) / model['stds'][c]
    # insert column of 1's
    Z1 = np.insert(Z, 0, 1, axis=1)
    output = np.dot(Z1,model['w'])
    return output
```

The `rmse` function takes the `predict` matrix produced by `use`, and the targets, `T`, as inputs. It then calculates the root mean square error and returns that.

```python
def rmse(predict,T):
    """Returns root mean square error between predict and T"""
    # T does not need standardization it seems
    error = predict - T
    sError = error **2
    msError = np.mean(sError, axis=0)
    rmse = np.sqrt(msError)
    return rmse
```

The `partitionKFolds` function takes 5 inputs. It takes in the input attributes, `X`, the targets, `T`, the number of folds, a boolean stating whether or not to shuffle the samples, and an integer for the number of partitions being used. If `shuffle` is true, the samples will be shuffled around in order to randomize the partition samples each time it is ran. Generally the default of `nPartitions = 3` will be used. The function will then yield `X` and `T` matrices for training, validation, and testing samples based on the number of folds.

```python
 def partitionKFolds(X,T,nFolds,shuffle=False,nPartitions=3):
    '''Usage: for Xtrain,Ttrain,Xval,Tval,Xtest,Ttext in partitionKFolds(X,T,5):'''
    # Randomly arrange row indices
    rowIndices = np.arange(X.shape[0])
    if shuffle:
        np.random.shuffle(rowIndices)
    # Calculate number of samples in each of the nFolds folds
    nSamples = X.shape[0]
    nEach = int(nSamples / nFolds)
    if nEach == 0:
        raise ValueError("partitionKFolds: Number of samples in each fold is 0.")
    # Calculate the starting and stopping row index for each fold.
    # Store in startsStops as list of (start,stop) pairs
    starts = np.arange(0,nEach*nFolds,nEach)
    stops = starts + nEach
    stops[-1] = nSamples
    startsStops = list(zip(starts,stops))
    # Repeat with testFold taking each single fold, one at a time
    for testFold in range(nFolds):
        if nPartitions == 3:
            # Repeat with validateFold taking each single fold, except for the testFold
            for validateFold in range(nFolds):
                if testFold == validateFold:
                    continue
                # trainFolds are all remaining folds, after selecting test and validate folds
                trainFolds = np.setdiff1d(range(nFolds), [testFold,validateFold])
                # Construct Xtrain and Ttrain by collecting rows for all trainFolds
                rows = []
                for tf in trainFolds:
                    a,b = startsStops[tf]                
                    rows += rowIndices[a:b].tolist()
                global Xtrain
                global Ttrain
                Xtrain = X[rows,:]
                Ttrain = T[rows,:]
                # Construct Xvalidate and Tvalidate
                a,b = startsStops[validateFold]
                rows = rowIndices[a:b]
                Xvalidate = X[rows,:]
                Tvalidate = T[rows,:]
                # Construct Xtest and Ttest
                a,b = startsStops[testFold]
                rows = rowIndices[a:b]
                global Xtest
                global Ttest
                Xtest = X[rows,:]
                Ttest = T[rows,:]
                # Return partition matrices, then suspend until called again.
                yield Xtrain,Ttrain,Xvalidate,Tvalidate,Xtest,Ttest,testFold
        else:
            # trainFolds are all remaining folds, after selecting test and validate folds
            trainFolds = np.setdiff1d(range(nFolds), [testFold])
            # Construct Xtrain and Ttrain by collecting rows for all trainFolds
            rows = []
            for tf in trainFolds:
                a,b = startsStops[tf]                
                rows += rowIndices[a:b].tolist()
            Xtrain = X[rows,:]
            Ttrain = T[rows,:]
            # Construct Xtest and Ttest
            a,b = startsStops[testFold]
            rows = rowIndices[a:b]
            Xtest = X[rows,:]
            Ttest = T[rows,:]
            # Return partition matrices, then suspend until called again.
            yield Xtrain,Ttrain,Xtest,Ttest,testFold
```

The `multipleLambdas` function takes `X`, `T`, the number of folds, and a range of values for $\lambdav$ as inputs. It then uses the 4 previously defined functions directly above to return a `bestResults` array that includes the fold index, the best value of $\lambdav$ for that fold, and the three errors: training, validation and testing set errors.

```python
def multipleLambdas(X, T, nFolds, lambdas=[0]):
    """Finds and returns best results of ridge regression across multiple folds and values of lambda"""
    foldCount = 0
    results = []
    for Xtrain, Ttrain, Xval, Tval, Xtest, Ttest, _ in partitionKFolds(X, T, nFolds, True):
        for lamb in lambdas:
            model = train(Xtrain, Ttrain, lamb)
            # find prediction and rmse with training values
            trainPredict = use(model,Xtrain)
            rmseTrain = rmse(trainPredict, Ttrain)
            # find prediction and rmse with validation values
            valPredict = use(model,Xval)
            rmseVal = rmse(valPredict, Tval)
            # find prediction and rmse with test values
            testPredict = use(model,Xtest)
            rmseTest = rmse(testPredict, Ttest)
            results.append([foldCount, lamb, rmseTrain, rmseVal, rmseTest])
        foldCount+= 1
    results = np.array(results)
    bestResults = []
    for fold in range(foldCount):
        rows = results[results[:,0] == fold, :]
        minValErr = min(rows[:,3])
        minRow = rows[rows[:,3] == minValErr, :]
        bestResults.append(minRow[0])
    return np.array(bestResults)
```

# Data

I picked a dataset from the UCI Machine Learning Repository. I decided to work with the [Airfoil Self-Noise Data Set.](http://archive.ics.uci.edu/ml/datasets/Airfoil+Self-Noise#)

The attributes are as follows:
    1. Frequency in units of Hertz. (Hz)
    2. Angle of attack in degrees.
    3. Chord length in meters. (m)
    4. Free-stream velocity in meters per second. (m/s)
    5. Suction side displacement thickness in meters. (m)
    6. Scaled sound pressure level in decibels. (dB)

Maintaining the same order as above, descriptions of the attributes are as follows:
     1. A measure of how often a vortex is shed from the airfoil. 
     2. The angle that the airfoil takes against the free stream velocity, relative to the horizontal. 
     3. Straight line measurement between the leading and trailing edges of the airfoil. 
     4. The speed of the air in the wind tunnel. 
     5. The thickness of the upper surface of the airfoil. 
     6. A measurement of how audible the airfoil's self-noise is.

```python
# I pulled this from the lecture notes and tweaked it slightly to work with my dataset
def makeAirfoilData(filename='airfoil_self_noise.dat'):
    def missingIsNan(s):
        return np.nan if s == b'?' else float(s)
    data = np.loadtxt(filename, usecols=range(6), converters={3: missingIsNan})
    #print("Read",data.shape[0],"rows and",data.shape[1],"columns from",filename)
    goodRowsMask = np.isnan(data).sum(axis=1) == 0
    data = data[goodRowsMask,:]
    #print("After removing rows containing question marks, data has",data.shape[0],"rows and",data.shape[1],"columns.")
    X = data[:,0:5]
    T = data[:,5:]
    Xnames =  ['Frequency','Angle of Attack','Chord Length','Free Stream Velocity','Suction Side Displacement']
    Tname = 'Scaled Sound Pressure Level'
    return X,T,Xnames,Tname
```

```python
X,T,Xnames,Tname = makeAirfoilData()
```

```python
plt.figure(figsize=(10,10))
nrow,ncol = X.shape
for c in range(ncol):
    plt.subplot(3,3, c+1)
    plt.plot(X[:,c])
    plt.ylabel(Xnames[c])
    plt.xticks(np.arange(0, 1600, 250))
plt.tight_layout()
```

It looks as though testing was done with intervals of varying angle of attack, free-stream velocity, and suction side displacement thickness. Chord length was decreasing as the study went on, excluding the final chord length used of .10 meters. Frequency looks to have went up and down in intervals as well, with an overall trend of reduced frequency. The sound seems to have a slightly upward trend until it dips down in the last hundred or so samples. 

The first 5 attributes are being used to predict scaled sound pressure.

```python
plt.figure(figsize=(10,10))
nrow,ncol = X.shape
for c in range(ncol):
    plt.subplot(3,3, c+1)
    plt.plot(X[:,c],T, 'o')
    plt.ylabel(Tname)
    plt.xlabel(Xnames[c])
plt.tight_layout()
```

With the control variables constantly moving up and down in intervals, it is difficult to see much correlation between any one input and the self-sound of the airfoil. One thing to notice, however, is that the sound tends to be louder when frequency takes on smaller values. Angle of attack looks to maximize the measured decibels around 7-8 degrees, while an increase in suction side displacement seems to have an overall inverse relationship with the sound levels.
