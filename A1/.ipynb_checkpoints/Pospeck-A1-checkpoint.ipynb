## Overview

Linear regression will be used on a [data set](http://archive.ics.uci.edu/ml/datasets/Airfoil+Self-Noise#) to try and predict how loud an airfoil is based on 5 other input attributes. A *train* function will be used to create a model from the airfoil data set, after it is standardized. The *use* function will use the model to create a predicted output matrix. After that, the *rmse* function will compare the predicted output against the actual output and report the root-mean-square-error. At the end, any relatively insignificant weights will be removed from the linear model and the whole process will be run again to see how predictions change.

## Imports

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

## Method

For this assignment, the following functions will be used:

  * ```model = train(X,T)```
  * ```predict = use(model,X)```
  * ```error = rmse(predict,T)```

The model, ```w```, and the data targets, ```T```, are column vectors. The input data, ```X```, will be a matrix. All 3 are defined below:

$$
    \begin{align*}
        T &= \begin{bmatrix}
          t_1 \\ t_2 \\ \vdots \\ t_N
        \end{bmatrix} \\
        X &= \begin{bmatrix}
          x_{1,0} & x_{1,1} & x_{1,2} & \dotsc & x_{1,D} \\
          x_{2,0} & x_{2,1} & x_{2,2} & \dotsc & x_{2,D} \\
          \vdots \\
          x_{N,0} & x_{N,1} & x_{N,2} & \dotsc & x_{N,D}
        \end{bmatrix}\\
        \wv &= \begin{bmatrix} w_0 \\ w_1 \\ \vdots \\ w_D \end{bmatrix}
      \end{align*}
$$  
    
    where:
$$
    \begin{align*}
    \wv &= (\Xv^T \Xv)^{-1} \Xv^T \Tv
    \end{align*}
$$

# Functions

````python
"""Standardizes the inputs, X, before calculating the means, standard deviations, 
and weights which are all returned in a dictionary"""
def train(X,T):
    mean = np.mean(X,axis=0)
    std = np.std(X,axis=0)
        # below for loop standardizes the inputs before calculating w
    cols = X.shape[1]
    Z = np.empty(X.shape)
    for c in range(cols):
        Z[:,c] = (X[:,c]-mean[c]) / std[c]
    Z1 = np.insert(Z, 0, 1, axis=1)
    w = np.linalg.lstsq(np.dot(Z1.T,Z1),np.dot(Z1.T,T))[0]
    return {'means': mean, 'stds': std, 'w': w}
````

````python
"""Uses the returned dictionary from the train function in order to calculate 
and return the predicted matrix"""
def use(model,X):
    Z = np.empty(X.shape)
    cols = X.shape[1]
    for c in range(cols):
        Z[:,c] = (X[:,c] - model['means'][c]) / model['stds'][c]
    Z1 = np.insert(Z, 0, 1, axis=1)
    output = np.dot(Z1,model['w'])
    return output
````

````python
"""Returns root mean square error between predict and T"""
def rmse(predict,T):
    error = predict - T
    sError = error **2
    msError = np.mean(sError, axis=0)
    rmse = np.sqrt(msError)
    return rmse
````

## Data

I picked a dataset from the UCI Machine Learning Repository. I decided to work with the [Airfoil Self-Noise Data Set.](http://archive.ics.uci.edu/ml/datasets/Airfoil+Self-Noise#)

The attributes are as follows:
    1. Frequency in units of Hertz. (Hz)
    2. Angle of attack in degrees.
    3. Chord length in meters. (m)
    4. Free-stream velocity in meters per second. (m/s)
    5. Suction side displacement thickness in meters. (m)
    6. Scaled sound pressure level in decibels. (dB)

Maintaining the same order as above, descriptions of the attributes are as follows:
     1. A measure of how often a vortex is shed from the airfoil. 
     2. The angle that the airfoil takes against the free stream velocity, relative to the horizontal. 
     3. Straight line measurement between the leading and trailing edges of the airfoil. 
     4. The speed of the air in the wind tunnel. 
     5. The thickness of the upper surface of the airfoil. 
     6. A measurement of how audible the airfoil's self-noise is.

Start by reading in the data and ensuring there are 6 columns: 1 for each attribute

````python
data = np.loadtxt('airfoil_self_noise.dat', usecols=range(6))
data.shape
````

Looks good. Now to setup the labels for each attribute and plot them all individually.

````python
labels = ['Frequency', 'Angle of Attack', 'Chord Length', 'Free-stream Velocity', 
          'Suction Side Displacement Thickness', 'Scaled Sound Pressure']
````

````python
plt.figure(figsize=(10,10))
rows, cols=data.shape
for c in range(cols):
    plt.subplot(3,3, c+1)
    plt.plot(data[:,c])
    plt.ylabel(labels[c])
        # formatting space around the figures
    plt.subplots_adjust(right=1.4)
    plt.subplots_adjust(top=1.4)
````

It looks as though testing was done with intervals of varying angle of attack, free-stream velocity, and suction side displacement thickness. Chord length was decreasing as the study went on, excluding the final chord length used of .10 meters. Frequency looks to have went up and down in intervals as well, with an overall trend of reduced frequency. The sound seems to have a slightly upward trend until it dips down in the last hundred or so samples. 

Now I'll use the first 5 attributes to predict scaled sound pressure. I'll make the target column vector, T, of size 1503x1. T will be filled with the values of scaled sound pressure. The input matrix, X, will be 1503x5 and will contain the other 5 attributes.

````python
T = data[:,5:]
````

````python
X = data[:,0:5]
````

````python
Tlabels = labels[5:]
Xlabels = labels[0:5]
Tlabels, Xlabels
````

Now that the input and output labels are separated apropriately, it's time to plot the output vs. all of the inputs individually and see what they look like.

````python
plt.figure(figsize=(10,10))
for c in range(X.shape[1]):
    plt.subplot(3,3, c+1)
    plt.plot(X[:,c], T, 'o', alpha=0.5)
    plt.ylabel(Tlabels[0])
    plt.xlabel(Xlabels[c])
        # formatting space around the figures
    plt.subplots_adjust(right=1.4)
    plt.subplots_adjust(top=1.4)
````

With the control variables constantly moving up and down in intervals, it is difficult to see much correlation between any one input and the self-sound of the airfoil. One thing to notice, however, is that the sound tends to be louder when frequency takes on smaller values. Angle of attack looks to maximize the measured decibels around 7-8 degrees, while an increase in suction side displacement seems to have an overall inverse relationship with the sound levels.
